# HECHOS - HOMICIDIOS

```python
df = pd.read_excel('../Data/excel/homicidios.xlsx',['HECHOS','VICTIMAS'])
```

```python
df_hechos = df['HECHOS']
```

```python
df_hechos.info()
```

```python
df_hechos.isnull().sum()
```

se observan muy pocas columnas con campos en null, procedemos a eliminar columnas donde se hace rebundante la info y renombrar las que necesitamos

```python
columnas = ['AAAA', 'MM', 'DD','HH','PARTICIPANTES','Calle','Cruce','Dirección Normalizada']
df_hechos = df_hechos.drop(columnas, axis=1)
```

```python
renombrar = {'ID': 'Id_hechos',
             'N_VICTIMAS': 'nro_victimas',
             'FECHA': 'fecha',
             'HORA': 'hora',
             'LUGAR_DEL_HECHO': 'lugar_hecho',
             'TIPO_DE_CALLE': 'tipo_calle',
             'Altura': 'altura',
             'COMUNA': 'comuna',
             'XY (CABA)': 'geocodificacion_CABA',
             'pos x': 'longitud',
             'pos y	': 'latutid',
             'VICTIMA': 'vehiculo_victima',
             'ACUSADO': 'vehiculo_acusado'}

df_hechos = df_hechos.rename(columns=renombrar)
df_hechos = df_hechos.rename(columns={'pos y': 'latitud'})
```

corremos una funcion a nuestradata para que sea mas legible

```python
def transformar_texto(columna):
    if pd.notnull(columna):
        return columna.strip().lower().capitalize().lower().capitalize()
    else:
        return columna 

def transformar_dataframe(df, columnas_a_transformar):
    for columna in columnas_a_transformar:
        df[columna] = df[columna].apply(transformar_texto)
```

```python
columnas = ['lugar_hecho', 'tipo_calle','vehiculo_victima','vehiculo_acusado']
transformar_dataframe(df_hechos, columnas)
```

```python
df_hechos.sample(5)
```

recordamos que tenemos 567 valores como nulos en nuestra columna altura,  procedemos a castearla para dejar en tipo entero
