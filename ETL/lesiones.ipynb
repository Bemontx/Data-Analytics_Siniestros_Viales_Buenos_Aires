```python
import pandas as pd
import numpy as np
```

### EXCEL LESIONES - HECHOS

```python
df = pd.read_excel('../Data/excel/lesiones.xlsx',['HECHOS','VICTIMAS'])
```

```python
df_hechos = df['HECHOS']
```

```python
df_hechos.info()
```

```python
df_hechos.isnull().sum()
```

```python
columnas = ['aaaa', 'mm', 'mm','franja_hora','otra_direccion','calle','cruce','moto','auto','transporte_publico','camion','ciclista']
df_hechos = df_hechos.drop(columnas, axis=1)
```

```python
renombrar = {'id': 'Id_hechos',
             'n_victimas': 'nro_victimas',
             'direccion_normalizada': 'lugar_hecho',
             'victima': 'vehiculo_victima',
             'acusado': 'vehiculo_acusado'}

df_hechos = df_hechos.rename(columns=renombrar)
df_hechos.rename(columns={'latutid': 'latitud'}, inplace=True)
```

```python
df_hechos.sample(5)
```

```python
def transformar_texto(columna):
    if pd.notnull(columna):
        return columna.strip().lower().capitalize().lower().capitalize()
    else:
        return columna 

def transformar_dataframe(df, columnas_a_transformar):
    for columna in columnas_a_transformar:
        df[columna] = df[columna].apply(transformar_texto)
```

```python
columnas = ['lugar_hecho', 'tipo_calle','vehiculo_victima','vehiculo_acusado','participantes','gravedad']
transformar_dataframe(df_hechos, columnas)
```

```python
df_hechos['gravedad'].unique()
```

```python
df_hechos['gravedad'] = df_hechos['gravedad'].replace({'Sd': 'Leve','GRAVE': 'Grave', 'grave': 'Grave', 'FATAL': 'Fatal'})
print(df_hechos['gravedad'].unique())
```

```python
df_hechos.to_csv('lesiones_hechos.csv', index=False)
```

EXCEL LESIONES - VICTIMAS

```python
df_victimas = df['VICTIMAS']
```

```python
df_victimas.info()
```

```python
df_victimas.isnull().sum()
```

```python
df_victimas.head()
```

```python
columnas = ['AAA', 'MM', 'DD','FECHA ','GRAVEDAD']
df_victimas = df_victimas.drop(columnas, axis=1)
```

```python
columnas = ['SEXO','VEHICULO_VICTIMA']
transformar_dataframe(df_victimas, columnas)
```

```python
renombrar = {'ID hecho': 'Id_hechos',
             'FECHA': 'fecha',
             'VEHICULO_VICTIMA': 'victima',
             'SEXO': 'sexo',
             'EDAD_VICTIMA': 'edad'}
df_victimas = df_victimas.rename(columns=renombrar)
```

```python
df_victimas.head()
```

### MERGE LESIONES HECHOS - VICTIMAS

```python
df_1 = pd.merge(df_hechos, df_victimas, on='Id_hechos', how='inner')
```

```python
df_1.info()
```

```python
df_1 = df_1.drop(['dd'], axis=1, errors='ignore')
df_1.rename(columns=lambda x: x.strip(), inplace=True)
df_1.rename(columns={'latutid': 'latitud'}, inplace=True)
```
